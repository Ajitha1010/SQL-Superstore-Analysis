```python
df = pd.read_csv(r"C:\Users\WIN 10\Documents\SQL\Pythonlearning\superstore_final_dataset.csv",encoding="latin1")
df=df.dropna(subset=['Postal_Code'])
df.isnull().sum()
```

```python
df['Order_Date']=pd.to_datetime(df['Order_Date'], dayfirst=True)
df['Ship_Date']=pd.to_datetime(df['Ship_Date'], dayfirst=True)
df.info()
```

```python
df.rename(columns={'Customer_Name':'Customer','Postal_Code': 'Zip_code'})
```

```python
import matplotlib.pyplot as plt

df.hist('Sales')
```

```python
category_sales= df.groupby('Category')['Sales'].sum()
category_sales.plot(x='Category',y='Sales',kind='bar')
```

```python
city_sales= df.groupby('City')['Sales'].sum().sort_values(ascending=False)
city_sales.head(10).plot(kind='bar')
```
